```python
import pandas as pd
import json
import datetime as dt
df=pd.read_json("C:/Users/Dileep Sathya/OneDrive/Desktop/Stock_AI_2.0/artifacts/hist_data.json")
df['year']=df['date'].dt.year
df
```

```python
df=df[df['year']==2025]
df
```

```python
df['prev_day_open'] = df.groupby('symbol')['open'].shift(1)
df['prev_day_high'] = df.groupby('symbol')['high'].shift(1)
df['prev_day_low'] = df.groupby('symbol')['low'].shift(1)
df['prev_day_close'] = df.groupby('symbol')['close'].shift(1)
```

```python
df=df.dropna()
df
```

```python
# Daily range % (High - Low relative to Low of the same day)
df['prev_day_range'] = abs((df['prev_day_high'] - df['prev_day_low']) / df['prev_day_low']) * 100

# Body range % (Close - Open relative to Open of the same day)
df['prev_day_body_range'] = abs((df['prev_day_close'] - df['prev_day_open']) / df['prev_day_open']) * 100


df
```

```python


# Calculate 25th percentiles (Q1)
range_25 = df['prev_day_range'].quantile(0.25)
body_range_25 = df['prev_day_body_range'].quantile(0.25)

# Apply the condition
df['eligibility'] = (df['prev_day_range'] < range_25) & (df['prev_day_body_range'] < body_range_25) 
df['range'] = abs((df['high'] - df['low']) / df['low']) * 100

# Body range % (Close - Open relative to Open of the same day)
df['body_range'] = abs((df['close'] - df['open']) / df['open']) * 100
df=df.drop(columns=['open','high','low','close','volume','year','prev_day_high','prev_day_low','prev_day_close','prev_day_open'])
df
```
