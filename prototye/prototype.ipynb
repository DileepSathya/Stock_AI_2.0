```python
import pandas as pd



# Convert 'date' to datetime if not already
df['date'] = pd.to_datetime(df['date'])

# Sort by symbol and date
df = df.sort_values(by=['symbol', 'date'])

# Create previous day features by group
df['prev_day_open'] = df.groupby('symbol')['open'].shift(1)
df['prev_day_high'] = df.groupby('symbol')['high'].shift(1)
df['prev_day_low'] = df.groupby('symbol')['low'].shift(1)
df['prev_day_close'] = df.groupby('symbol')['close'].shift(1)
df['prev_day_volume'] = df.groupby('symbol')['volume'].shift(1)




df=df.dropna()
df=df.sort_index()
df
```

```python
df['close'].describe()
```

```python
import numpy as np

def get_ltp_range(value, bucket_size=500):
    lower = (value // bucket_size) * bucket_size
    upper = lower + bucket_size
    return f"{int(lower)}-{int(upper)}"

df['ltp_range'] = df['close'].apply(get_ltp_range)

df
```

```python
df['ltp_range'].describe()
```

```python
import datetime as dt
df['month']=df['date'].dt.month
df['year']=df['date'].dt.year
df
```

```python
df['week']=df['date'].dt.weekday
df
```

```python
# 5 EMA
df['ema_5'] = df.groupby('symbol')['prev_day_close'].transform(lambda x: x.ewm(span=5, adjust=False).mean()).round(2)

# 20 EMA
df['ema_20'] = df.groupby('symbol')['prev_day_close'].transform(lambda x: x.ewm(span=20, adjust=False).mean()).round(2)

# 50 EMA
df['ema_50'] = df.groupby('symbol')['prev_day_close'].transform(lambda x: x.ewm(span=50, adjust=False).mean()).round(2)
df
```

```python
df['prev_prev_day_close'] = df.groupby('symbol')['close'].shift(2)
df['prev_day_pct_change'] = ((df['prev_day_close'] - df['prev_prev_day_close']) / df['prev_prev_day_close']) * 100
df['prev_day_pct_change'] = df['prev_day_pct_change'].round(2)
df
```
